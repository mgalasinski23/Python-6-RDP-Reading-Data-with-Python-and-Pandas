```python
import numpy as np
import pandas as pd
import json
```

```python
artists_file = open('files/artists.json')
artists_lod = json.load(artists_file) #list of dicts
print(type(artists_lod))
artists_lod
```

```python
artists_df = pd.json_normalize(artists_lod)
artists_df_2 = pd.DataFrame.from_dict(artists_lod)

artists_df_2==artists_df
```

```python
artists_bio_df = pd.json_normalize(artists_lod,record_path='bio',meta='name')
artists_bio_df.head()
```

## Important note
Basically you want to get it as a <u>list</u> of <u>dictionaries</u> - after that everything else follows.

So the "from_dict" / json_normalize method is actually a bit misleading.

It can take in a dictionary of dictionaries, but it also takes in lists! The dictionaries are just the list items.
