```python
import numpy as np
import pandas as pd

data_url = 'https://github.com/ine-rmotr-projects/project-files/files/4086772/playstore.xlsx'
```

```python
playstore_file = pd.ExcelFile(data_url)
sheet_names = playstore_file.sheet_names
print(playstore_file)
```

```python
playstore_df = playstore_file.parse(sheet_name = sheet_names[0],
                                    usecols=["App","Rating","Installs","Genres","Last_Updated"],
                                    parse_dates=[-1])


```

```python
playstore_df.head()
```

```python
filtered_df = playstore_df.sort_values("Rating",ascending=False)[:25]
filtered_df
```
