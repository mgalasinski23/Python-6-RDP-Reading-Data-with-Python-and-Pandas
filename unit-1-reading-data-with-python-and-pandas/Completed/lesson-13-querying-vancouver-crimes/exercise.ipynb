```python
import numpy as np
import pandas as pd
import sqlite3 as sql3
```

```python
conn = sql3.connect(':memory:')
cur = conn.cursor()

sql_script_text=open('files/van_crime_2003.sql', 'r').read()

cur.executescript(sql_script_text)
```

```python
crimes = pd.read_sql("SELECT * FROM van_crimes",conn)
crimes.head()
```

```python
sql_query="""
    SELECT TYPE, MONTH, DAY, NEIGHBOURHOOD 
    FROM van_crimes
    WHERE NEIGHBOURHOOD='Stanley Park' OR NEIGHBOURHOOD='West End';
"""

crimes_filt=pd.read_sql(sql_query,conn)
crimes_filt.head(25)
```

```python
crime_types=crimes['TYPE'].value_counts()
crime_types.head()
```
